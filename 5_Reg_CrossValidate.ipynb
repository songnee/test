## 교차 검증 (Cross Validation)
![](https://velog.velcdn.com/images/newnew_daddy/post/9d55a412-b773-46a2-ba61-d36c87de648f/image.png)
#### 학습중인 모델의 성능을 평가하는 방법으로 고정된 검증 데이터셋에만 과적합 되어버리지 않도록 하는 방법
> 1. 데이터를 5개의 폴드로 나눕니다. 각 폴드는 20개의 데이터를 포함합니다.
> 2. 첫 번째 폴드를 **테스트 데이터**로 사용하고, 나머지 80개는 **학습**에 사용하여 모델을 학습합니다. 그런 다음 첫 번째 폴드에 대해 예측을 수행합니다.
> 3. 이 과정을 5번 반복하여, 각 폴드가 한 번씩 테스트 데이터로 사용되며, 해당 폴드에 대한 예측값이 생성됩니다.
> 4. **결과적으로 최종 모델은 생성되지 않고**, 각 데이터 포인트가 테스트 세트로 있을 때 예측된 값이 반환됩니다.

#### cross_val_score
- 특정 평가 지표를 사용하여 교차 검증을 수행하고, 각 교차 검증 반복에서 얻은 점수만 반환

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

# 회귀용 데이터 생성
X, y = make_regression(
    n_samples=5643,
    n_features=6,
    n_targets=1,
    noise=50
    )

dd = pd.DataFrame(X, columns=[f'feature{i}' for i in range(1,7)])
dd['target'] = y
dd
```

```python
df = pd.read_csv("dataset/random_reg.csv")

X = df[df.columns[:-1]]
y = df['target'].to_frame()
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
from sklearn.model_selection import cross_val_score

## 교차 검증 수행
## scoring 지표 확인 -> https://scikit-learn.org/stable/modules/model_evaluation.html#scoring-parameter
scores = cross_val_score(lr, X, y, cv=5, scoring='r2')

print(f"Cross-Validation Scores: {scores}")
print(f"Mean Score: {scores.mean()}")
```

```python
from sklearn.datasets import load_diabetes

dia = load_diabetes()
X = dia.data
y = dia.target

X.shape, y.shape
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

X = ss.fit_transform(X)
```

```python
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor

lr = LinearRegression()
sgd = SGDRegressor()
svr = SVR()
tree = DecisionTreeRegressor(max_depth=3)
```

```python
models = [lr, sgd, svr, tree]

for model in models:
    scores = cross_val_score(model, X, y, cv=5, scoring='r2')

    print(f"Cross-Validation Scores: {scores}")
    print(f"Mean Score: {scores.mean()}")
    print()
```

```python
scores = cross_val_score(lr, X, y, cv=5, scoring='r2')

print(f"Cross-Validation Scores: {scores}")
print(f"Mean Score: {scores.mean()}")
```

#### cross_validate
- 여러 개의 평가 지표를 사용할 수 있음
- 훈련 시간(fit_time), 예측 시간(score_time) 등을 포함한 다양한 정보를 반환
- 더 유연하고 다양한 정보를 제공

```python
from sklearn.model_selection import StratifiedKFold, KFold
from sklearn.model_selection import cross_validate

spl = KFold(n_splits=4, shuffle=True, random_state=20)

cv = cross_validate(
    lr,
    X,
    y,
    cv=spl,
    scoring=['r2', 'neg_root_mean_squared_error'],
    return_train_score=True
    )


indices = [f"fold_{i}" for i in range(1,5)]

pd.DataFrame(cv, index=indices)
```

```python
cv = cross_validate(
    lr,
    X,
    y,
    cv=5,
    scoring=['r2', 'neg_root_mean_squared_error'],
    return_train_score=True
    )

indices = [f"fold_{i}" for i in range(1,6)]

pd.DataFrame(cv, index=indices)
```

#### cross_val_predict
- 교차 검증을 통해 모델을 학습시키고, 학습된 모델로 테스트 세트의 예측값을 반환
- 예측값만 반환하며, 점수나 시간을 제공하지 않음

```python
from sklearn.model_selection import cross_val_predict

y_pred = cross_val_predict(lr, X, y, cv=5)
```
